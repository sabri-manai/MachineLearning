## Q7


```python
import math
C = 50
H = 30
D = str (input ("enter D :"))
lista = D.split(",")
listaokhra= [] 
for i in lista:
    k = int(i)
    Q = math.sqrt ((2 * C * k)/H)
    listaokhra.append(str(round(Q)))
print(listaokhra)
print(",".join(listaokhra))
```
